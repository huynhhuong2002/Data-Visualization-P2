### <span style='color:#2B9C15'>12. Kh√°m ph√° m·ªëi quan h·ªá t·ª´ tr∆∞·ªùng d·ªØ li·ªáu job_title v√† work year </span>

üëâ ƒê·∫ßu ti√™n, ta s·∫Ω ph√¢n lo·∫°i c·ªôt job title th√†nh 4 lo·∫°i c√¥ng vi·ªác ƒë√≥ l√†: Machine Learning Engineer, Data Scientist, Data Architect, Data Engineer, Data Analyst. Nh·ªØng c√¥ng vi·ªác chi·∫øm s·ªë √≠t c√≤n l·∫°i s·∫Ω ƒë∆∞·ª£c ph·∫ßn v√†o lo·∫°i Other.

```python
JOB_TYPES = {
    "Machine Learning Engineer": ["machine learning", "autonomous", "ai", "mlops", "computer vision", "deep learning", "ml", "nlp", "research engineer"],
    "Data Scientist": ["science", "scientist"],
    "Data Architect": ["architect"],
    "Data Engineer": ["data engineer", "etl", "engineer"],
    "Data Analyst": ["analyst", "analytics", "bi", "business intelligence"],
}

def map_job_title(title):
    title = title.lower()
    for job_type, keywords in JOB_TYPES.items():
        if any(keyword in title for keyword in keywords):
            return job_type
    return "Other"

salaries_df["Job_Type"] = [map_job_title(title) for title in salaries_df["job_title"]]
```

üëâTa tr·ª±c quan b·∫±ng bi·ªÉu ƒë·ªì ƒë·ªÉ xem lo·∫°i c√¥ng vi·ªác n√†o t·∫≠p trung nhi·ªÅu ngu·ªìn nh√¢n l·ª±c:

```python
salaries_df.Job_Type.value_counts()
sns.set(rc={'figure.figsize':(14,5)})

ax=sns.countplot(data=salaries_df, x="Job_Type", color='aquamarine', saturation=0.1, order=salaries_df.Job_Type.value_counts(ascending=True).index)
for label in ax.containers:
    ax.bar_label(label)
    
patch_h = [patch.get_height() for patch in ax.patches]   
idx_tallest = np.argmax(patch_h)   
ax.patches[idx_tallest].set_facecolor('slateblue') 

plt.show()
```

+ Hi·ªÉu l√† t·ª∑ l·ªá s·ªë l∆∞·ª£ng nh√¢n vi√™n c√≥ th·ªÉ l√†m vi·ªác t·ª´ xa (remote) so v·ªõi t·ªïng s·ªë l∆∞·ª£ng nh√¢n vi√™n.
